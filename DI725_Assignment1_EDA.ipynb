```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

train_path = '/content/drive/MyDrive/customer_service/train.csv'
test_path = '/content/drive/MyDrive/customer_service/test.csv'
train_df = pd.read_csv(train_path, delimiter=",")
test_df = pd.read_csv(test_path, delimiter=",")

plt.figure(figsize=(8, 5))
sns.countplot(data=train_df, x='customer_sentiment', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution in Train Dataset')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=test_df, x='customer_sentiment', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution in Test Dataset')
plt.show()

#Issue Area (train)
plt.figure(figsize=(10, 5))
sns.countplot(data=train_df, x='issue_area', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Issue Areas (Train Dataset)')
plt.xticks(rotation=45)
plt.show()

#Issue Area (test)
plt.figure(figsize=(10, 5))
sns.countplot(data=test_df, x='issue_area', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Issue Areas (Test Dataset)')
plt.xticks(rotation=45)
plt.show()



#Issue Complexity (train)
plt.figure(figsize=(8, 5))
sns.countplot(data=train_df, x='issue_complexity', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Issue Complexity (Train Dataset)')
plt.show()

#Issue Complexity (test)
plt.figure(figsize=(8, 5))
sns.countplot(data=test_df, x='issue_complexity', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Issue Complexity (Test Dataset)')
plt.show()



#Agent Experience (train)
plt.figure(figsize=(8, 5))
sns.countplot(data=train_df, x='agent_experience_level', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Agent Experience Level (Train Dataset)')
plt.show()

#Agent Experience (test)
plt.figure(figsize=(8, 5))
sns.countplot(data=test_df, x='agent_experience_level', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Agent Experience Level (Test Dataset)')
plt.show()



#Product Category (train)
plt.figure(figsize=(10, 5))
sns.countplot(data=train_df, x='product_category', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Product Category (Train Dataset)')
plt.xticks(rotation=45)
plt.show()

#Product Category(test)
plt.figure(figsize=(10, 5))
sns.countplot(data=test_df, x='product_category', hue='customer_sentiment', palette='viridis')
plt.title('Sentiment Distribution across Product Category (Test Dataset)')
plt.xticks(rotation=45)
plt.show()
```
